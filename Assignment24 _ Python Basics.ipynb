# 1. What is the relationship between def statements and lambda expressions ?

    Both def and lambda are used as Functions in Python

```python
# def Example
def cube(n):
    return n**3

cube(3)
```

```python
# lambda Example
cube = lambda x:x**3
cube(3)
```

# 4. What are function annotations, and how are they used?

```python
# These are user-defined data that add some information to functions. 
# Annotations add information about the functions arguments and result. 
# They are optional and make no difference for the interpreter. 
# Theyre rather for us than the interpreter.
```

```python
def func(a,b): #Function without Anotation
    return a+b
```

```python
def func(a:int,b:int): #Function with Anotation
    return a+b
```

```python
def fun(a:int, b:list, c:str)-> 'list':
    return a+b+c
    
fun(1,2,3) 
# So, annotations are making no difference in the Final Result
```

# 5. What are recursive functions, and how are they used?

```python
# A recursive function is a function in code that refers to itself for execution.
```

```python
def factorial(x):
    if x ==1:
        return 1
    else:
        return (x*factorial(x-1))
factorial(5)
```
