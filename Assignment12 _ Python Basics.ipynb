# 5. What methods do you use to rotate a page?

```python
# page.rotate(90) # Rotate a page clockwise by increments of 90 degrees.
# page.rotate_clockwise
# page.rotateClockwise
# page.rotateCounterClockwise
```

```python
import PyPDF2

pdfIn = open("C://Users//Logins//Desktop//Heat Transfer IES GATE IAS 20 Years Question and Answers.pdf", 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfIn)
pdfWriter = PyPDF2.PdfFileWriter()

for pageNum in range(pdfReader.numPages):
    page = pdfReader.getPage(pageNum)
    page.rotateClockwise(90)
    pdfWriter.addPage(page)

pdfOut = open('rotated.pdf', 'wb')
pdfWriter.write(pdfOut)
pdfOut.close()
pdfIn.close()
```
